```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
#To read the dataset stored in the form of csv
lp=pd.read_csv('/content/iris.data.csv')
df=pd.DataFrame(lp)
print(df)
plt.style.use('ggplot')
#pd.set_option('max_columns',200)
```

```python
#rename columns
df=df.rename(columns={'value1':'V1','value2':'V2','value3':'V3','value4':'V4','Species':'species'})
```

```python
#dropping a column,axis=1 for a column
df.drop(['V1'],axis=1)
```

```python
#display column names
df.columns
```

```python
df.loc[df.duplicated(subset=['V1'])]
```

```python
#check for duplicate species name
'''
here since all the entries from row 1 to row 20 have the same values in the species column we can see the result is a consecutive display of the rows hen asked to display duplicated rows
'''
df.loc[df.duplicated(subset=['species'])].head(20)
```

```python
# all rows where the a specific species name is duplicated
df.query('species=="Iris-setosa"')
```

```python
#To display the duplication of a specific species upto a cerain row
df.query('species=="Iris-setosa"').head(20)
```

```python
#sum of all the rows where species names are duplicated are duplicated
df.duplicated(subset=['species']).sum()
```

```python
#values with no duplication in the species column
df.loc[~df.duplicated(subset=['species'])]
```

```python
#reducing no.of rows to unique values
df.loc[~df.duplicated(subset=['species'])].reset_index(drop=True).copy()
```
